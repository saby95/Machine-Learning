```
import pandas as pd
import numpy as np
import json
import networkx as nx
from itertools import permutations 
```

```
with open('Movie_Data.txt', 'rU') as f:
   data = f.readlines()
data = map(lambda x: x.rstrip(), data)
data_json_str = "[" + ",".join(data) + "]"
data_df = pd.read_json(data_json_str)
```

```
cast_list = data_df['cast'].values.tolist()
G=nx.Graph()
cast_dict = {}
```

```
n = 0
for cast in cast_list:
  node_list = []
  for actor in cast:
    if actor not in cast_dict.keys():
      cast_dict[actor] = n
      G.add_node(n)
      n = n+1
    node_list.append(cast_dict[actor])
  perm = permutations(node_list, 2)
  G.add_edges_from(perm)
```

```
def find_bacon_number(actor_name):
  if not actor_name in cast_dict.keys():
    print ("Actor name not found in list")
    return -1
  return (len(nx.shortest_path(G,source=cast_dict['Kevin Bacon'],target=cast_dict[actor_name])) -1)
```
