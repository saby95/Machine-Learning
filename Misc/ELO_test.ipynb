```
import keras
from keras.models import Sequential
from keras.layers import Dense, Activation
from sklearn.model_selection import train_test_split
from keras.utils import to_categorical
```

```
train_data = np.array(train[['card_id','feature_1','feature_2','feature_3']].values.tolist())
train_labels = np.array(train['target'].values.tolist())
```

```
model = Sequential()
model.add(Dense(8,activation='relu'))
model.add(Dense(5,activation='relu'))
model.add(Dense(1,activation='relu'))
```

```
model.compile('Adam',loss='mean_squared_error', metrics=['mean_squared_error'])
```

```
model.fit(train_data,train_labels,epochs=1,batch_size=64)
```
