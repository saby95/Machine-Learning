```python
import numpy
from numba import jit
```

```python
@jit 
def findCubeSums(N):
    start = -1000
    stop = 1000
    for i in range(start,stop+1):
        for j in range(start,stop+1):
            for k in range(start,stop+1):
                if ((i**3 + j**3 + k**3) == N):
                    print(i,j,k)            
```

```python
for val in range(50, 100+1):
    print("N = ",val)
    findCubeSums(val)
```
