```
def FindThreeNumbers(N):
  mem = set()
  for i in range(-1000, 1001):
        for j in range(i, 1001):
            k = np.cbrt(N - (i**3 + j**3))
            if k.is_integer() and k>=-1000 and k<= 1000 and ((i*j*k) not in mem):
                print(i,"\t",j,"\t", int(k)) 
                mem.add(i*j*k)
```

```
for n in range(50,101):
  print("N="+str(n))
  FindThreeNumbers(n)
```

```
from numba import jit
import numpy as np

n_list=np.arange(50,101)
a=np.arange(-1000,1001)
b=np.arange(-1000,1001)
c=np.arange(-1000,1001)
d=np.power(a,3)
e=np.power(b,3)
f=np.power(c,3)

for n in n_list:
    print("N = ",n)
    @jit(nopython=True, parallel=True)
    def go_fast(r,s,t,x,y,z):
        return_list=[]
        for i in range(len(r)):
            for j in range(len(s)):
                for k in range(len(t)): 
                    if (x[i]+y[j]+z[k]==n):
                        unsorted_list=[r[i],s[j],t[k]]
                        return_list.append(unsorted_list)
        return return_list

    returned_list=go_fast(a,b,c,d,e,f)
    
    if (len(returned_list)!=0):
        sorted_list=[]
        for elem in list(map(sorted,returned_list)):
            if elem not in sorted_list:
                sorted_list.append(elem)
    
        print(sorted_list)
```
