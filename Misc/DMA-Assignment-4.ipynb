```
import numpy as np
from math import sqrt

def density_connected(i, arr, k, dic, core_num):
  for j in range(len(arr)):
    if i == j:
      continue
    try:
      cluster_id[dic[j]]
      continue
    except(KeyError):
      pass
    if distance_points(arr[i],arr[j]) <= 2:
      cluster_id[dic[j]] = k
      if j in core_num:
        density_connected(j, arr, k, dic, core_num)
        
def distance_points(i,j):
  dist = (i[0] - j[0])**2  + (i[1] - j[1])**2
  dist = sqrt(dist)
  return dist

a = [5,8]
b = [10,8]
c = [11,8]
d = [6,7]
e = [10,7]
f = [12,7]
g = [13,7]
h = [5,6]
i = [10,6]
j = [13,6]
k = [6,5]
l = [9,4]
m = [11,5]
n = [14,6]
o = [15,5]
p = [2,4]
q = [3,4]
r = [5,4]
s = [6,4]
t = [7,4]
u = [15,4]
v = [3,3]
w = [7,3]
x = [8,2]

lsit = [a,b,c,d,e,f,g,h,i,j,k,l,m,n,o,p,q,r,s,t,u,v,w,x]

arr = np.array([a,b,c,d,e,f,g,h,i,j,k,l,m,n,o,p,q,r,s,t,u,v,w,x],dtype=int)
epi = 2
minpts = 3
core =[]
core_num = []

gh = 'a'
dic = {}
ddr = {}
k = 0
cluster_id = {}

for g in range(24):
  dic[g] = gh
  gh = chr(ord(gh)+1)
 
for i in range(len(arr)):
  kk = 0
  gg = []
  for j in range(len(arr)):
    if distance_points(arr[i],arr[j]) <= 2:
      kk+=1
      gg.append(dic[j])
      
  ddr[dic[i]] = gg
  if(kk>=minpts):
    core.append(dic[i])
    core_num.append(i)

print(core)
    
for i in range(len(arr)):
  try:
    cluster_id[dic[i]]
    continue
  except(KeyError):
    pass
  k = k+1
  cluster_id[dic[i]] = k
  density_connected(i, arr, k, dic, core_num)

#print(cluster_id)

cluster_proper = {}
for key,value in cluster_id.items():
  try:
    cluster_proper[value].append(key)
  except(KeyError):
    cluster_proper[value] = [key]

print(cluster_proper)
```
